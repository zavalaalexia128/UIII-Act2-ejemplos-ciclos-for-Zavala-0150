```python
limite = int(input("¿Hasta qué número quieres sumar?: ")) # 1. Recibe el límite del usuario
suma = 0                                                 # 2. Inicializa el acumulador en cero

for i in range(1, limite + 1):                           # 3. Itera desde 1 hasta el limite incluido [3]
    suma += i                                            # 4. Suma el valor actual de i al total
    print(f"Sumando {i}... Total parcial: {suma}")      # 5. Muestra el progreso de la suma

print(f"La suma total es: {suma}")                      # 6. Muestra el resultado final
```

```python
entrada = input("Ingresa nombres separados por comas: ") # 1. Recibe una cadena de texto
invitados = entrada.split(",")                         # 2. Convierte la cadena en una lista de nombres [4]

for nombre in invitados:                               # 3. Recorre cada elemento de la lista creada [3]
    nombre_limpio = nombre.strip()                     # 4. Elimina espacios extra alrededor del nombre [4]
    print(f"¡Bienvenido/a, {nombre_limpio}!")           # 5. Imprime un saludo personalizado para cada uno
```

```python
tabla_del = int(input("¿De qué número quieres la tabla?: ")) # 1. Recibe el número base

for i in range(1, 11):                                      # 2. Define el rango del 1 al 10 [5]
    resultado = tabla_del * i                               # 3. Realiza la operación aritmética [6]
    print(f"{tabla_del} x {i} = {resultado}")               # 4. Muestra la línea de la tabla calculada
```

```python
numeros = [9-15]                   # 1. Lista de números predefinida
buscado = int(input("Ingresa el número a buscar: "))    # 2. Recibe el objetivo del usuario

for n in numeros:                                       # 3. Inicia el recorrido secuencial [16]
    if n == buscado:                                    # 4. Compara el elemento actual con el buscado
        print(f"¡Número {buscado} encontrado!")         # 5. Informa del hallazgo
        break                                           # 6. Termina el ciclo prematuramente [7, 8]
else:                                                   # 7. Se ejecuta solo si el ciclo terminó sin break [7]
    print("El número no está en la lista.")             # 8. Informa si no se encontró nada
```

```python
precios = {"manzana": 1.2, "naranja": 0.8, "kiwi": 2.5} # 1. Diccionario de productos y precios [17]
maximo = float(input("¿Cuál es tu presupuesto máximo?: ")) # 2. Recibe el filtro de precio

print("Productos que puedes comprar:")
for fruta, precio in precios.items():                  # 3. Obtiene pares clave-valor del diccionario [17, 18]
    if precio <= maximo:                               # 4. Verifica si el precio entra en el presupuesto
        print(f"- {fruta}: ${precio:.2f}")              # 5. Muestra los productos que cumplen la condición
```
